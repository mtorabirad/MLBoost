```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_points = 1000
X = np.random.uniform(0, 1, num_points)
a_0, a_1, a_2 = 1, 2, 0.1
Y_true = a_0 + a_1*X + a_2*X**2


noise = np.random.normal(0, 0.2, num_points)
Y = Y_true + noise
```

```python
plt.scatter(X, Y)
plt.plot(X, Y_true, color='k')
```
